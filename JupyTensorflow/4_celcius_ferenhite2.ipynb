```python
import tensorflow as tf
```

```python
import numpy as np
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)
```

```python
celcius_q = np.array([-40,-10,0,8,15,22,38], dtype=float)
fahrenheit_a = np.array([-40,14,32,46,59,72,100],dtype=float)
```

```python
for i,c in enumerate(celcius_q):
    print('{} degrees of celcius is {} degree of fahreheit'.format(c,fahrenheit_a[i]))
```

```python
l0 = tf.keras.layers.Dense(units=4, input_shape=[1])
l1 = tf.keras.layers.Dense(units=4)
l2 = tf.keras.layers.Dense(units=1)
```

```python
model = tf.keras.Sequential([l0,l1,l2])
```

```python
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
history = model.fit(celcius_q,fahrenheit_a, epochs=500, verbose=True)
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Epochs')
plt.ylabel('loss')
plt.plot(history.history['loss'])
```

```python
model.predict([100.0])
```

```python
print('l0 weights: {}'.format(l0.get_weights()))
print('l1 weights: {}'.format(l1.get_weights()))
print('l2 weights: {}'.format(l2.get_weights()))
```
